# Reticle Python bindings: open, generate, render

This notebook opens a GDSII layout, lists the built-in generators, places a
guard ring into a cell, and renders that cell inline. Rendering needs a GPU
adapter; on a headless machine the render step degrades to a short note instead
of an image, and the notebook still runs top to bottom.

```python
import reticle_py as rp

print("reticle-py", rp.version())
for g in rp.generators():
    print(f"  {g['id']:<16} {g['title']}")
```

```python
# nbconvert runs with the working directory set to this notebook's folder, so
# the sample layout beside it opens by name.
doc = rp.Document.open("basic.gds")
print(doc)
print("summary:", doc.summary())
print("cells:", doc.cell_names())
```

```python
# Place a guard ring into the first cell. Empty params `{}` fall back to the
# generator's defaults, which are a working ring on their own.
cell = (doc.top_cells() or doc.cell_names())[0]
result = doc.place_generator(cell, "guard_ring", "{}")
print(f"placed guard_ring into {cell!r}: {result}")
print("shapes now:", len(doc.shapes(cell)))
```

```python
# Render the cell inline. `show` returns an IPython image when a GPU is present,
# otherwise it prints a note and returns None.
rp.show(doc, cell, width=640, height=480)
```

```python
# Save the modified document back to a GDSII file.
doc.save("with_guard_ring.gds")
print("wrote with_guard_ring.gds")
```
